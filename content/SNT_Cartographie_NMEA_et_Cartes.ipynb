## <u>I. Analyse de données satellite : trame NMEA</u>:

<img src="https://capytale2.ac-paris.fr/web/sites/default/files/2022/03-09/19-36-37/Logo.jpg" width="100">

Un récepteur GPS (votre smartphone) détermine par calcul sa position et transmet les données de géolocalisation sous forme d’une trame (une suite) de caractères : c’est la <b><u>trame NMEA</u></b> (National Marine Electronics Association). Un ordinateur décode cette trame et affiche éventuellement la carte correspondante. <br><br>

Chaque trame commence par le caractère ‘<b>$</b>’, suivi des lettres ‘<b>GP</b>’ (pour Global Positioning System), puis un groupe de 3 lettres pour identifier le type de trame. 

<u><b>Voici un exemple</b></u> : 
$GP<b>GGA</b>,073224.279,4234.4120,N,00240.9373,E,1,04,3.2,199.2,M,,,,0000*0E

-->Chaque donnée de cette trame est séparée par …une virgule 

Voici la signification de ces données (à connaître !) : 

<img src="https://capytale2.ac-paris.fr/web/sites/default/files/2022/03-09/19-36-37/nmea.jpg">

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pylab

G = nx.Graph()

G.add_edges_from([('A', 'B'),('E','H'),('H','I'), ('F','I'), ('G','I')], weight=40)
G.add_edges_from([('A','C')], weight=75)
G.add_edges_from([('C','F'), ('B','E')], weight=100)
G.add_edges_from([('A','D')], weight=90)
G.add_edges_from([('D','G')], weight=120)

edge_labels=dict([((u,v,),d['weight'])
                 for u,v,d in G.edges(data=True)])

pos=nx.circular_layout(G)
#commande pour faire des arêtes avec indication de poids
# commande qui utilise pylab
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels)
#commande pour tracer le graphe
nx.draw(G,pos, node_color = 'yellow',with_labels=True, node_size=1500,edge_color='black',edge_cmap=plt.cm.Reds)
pylab.show()

print("La distance la plus courte : ",(nx.dijkstra_path_length(G,'A','I')))
print(".. correspondant à ce trajet : ",(nx.dijkstra_path(G,'A','I')))
```
